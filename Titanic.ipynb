# **Read the CSV file using pandas**

```python
import pandas as pd
```

```python
titanic=pd.read_csv("https://raw.githubusercontent.com/HariharanR344/Titanic/refs/heads/main/Titanic-Dataset.csv")
titanic
```

# **Checking for Null values**

```python
titanic.isnull().sum()
```

```python
# Age
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())

# Embarked
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])

# Cabin
titanic.drop(columns=['Cabin'], inplace=True)


titanic
```

```python
titanic.isnull().sum()
```

# **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
titanic['Sex_encoded'] = le.fit_transform(titanic['Sex'])
titanic
```

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic
```

# **Feature and Target Selection**

```python
feature=titanic[['Pclass','Age','Sex_encoded','Fare','FamilySize']]
target=titanic['Survived']
```

```python
X = feature
y = target
```

```python
survival_count=y.value_counts()
survival_count
```

```python
y.value_counts().plot.pie(autopct='%.2f')
```

# **Spliting**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4
)
```

# **Under Sampling**

```python
! pip install -U imbalanced-learn
```

```python
from imblearn.under_sampling import RandomUnderSampler


rus = RandomUnderSampler()
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)
```

```python
survival_count=y_train_rus.value_counts()
survival_count
```

```python
ax = y_train_rus.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title("Under-sampling")
```

# **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train_rus)
X_test_scaled = scaler.transform(X_test)
```

# **Training**

```python
from sklearn.ensemble import RandomForestClassifier
dt=RandomForestClassifier(n_estimators=10)
dt.fit(X_train_scaled, y_train_rus)
```

# **Prediction**

```python
ypredicted=dt.predict(X_test_scaled)
print(ypredicted)
```

# **Classification Report**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, ypredicted))
```

# **Confusion Matrix**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, ypredicted)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
